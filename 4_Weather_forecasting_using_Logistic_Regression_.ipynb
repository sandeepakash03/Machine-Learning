```python
# finding up the numeric and categorical columns
numeric_cols=raw_df.select_dtypes(include=np.number).columns.tolist()
categorical_col=raw_df.select_dtypes('object').columns.tolist()
```

```python
from sklearn.impute import SimpleImputer
impute=SimpleImputer(strategy='mean')
raw_df[numeric_cols]=impute.fit_transform(raw_df[numeric_cols])
```

```python
from sklearn.impute import SimpleImputer
impute=SimpleImputer(strategy='most_frequent')
raw_df[categorical_col]=impute.fit_transform(raw_df[categorical_col])
```

```python
from sklearn.model_selection import train_test_split
train_val_df,test_df=train_test_split(raw_df,test_size=0.2,random_state=42)
train_df,val_df=train_test_split(train_val_df,test_size=30,random_state=42)
```

```python
sns.countplot(x=pd.to_datetime(raw_df['Date']).dt.year)
plt.title('No of rows per year')
plt.show()
```

```python
year=pd.to_datetime(raw_df['Date']).dt.year

train_df=raw_df[year<2015]
val_df=raw_df[year==2015]
test_data=raw_df[year>2015]
```

```python
input_cols=list(train_df.columns)[1:-1]
target_cols='RainTomorrow'
```

```python
train_inputs=train_df[input_cols].copy()
train_target=train_df[target_cols].copy()

val_inputs=val_df[input_cols].copy()
val_target=val_df[target_cols].copy()

test_inputs=test_df[input_cols].copy()
test_target=test_df[target_cols].copy()
```

```python
numeric_cols
```

```python
raw_df.isna().sum( )
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
train_inputs[numeric_cols]=scaler.fit_transform(train_inputs[numeric_cols])
val_inputs[numeric_cols]=scaler.fit_transform(val_inputs[numeric_cols])
test_inputs[numeric_cols]=scaler.fit_transform(test_inputs[numeric_cols])
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
```

```python
categorical_cols=categorical_col[:-1]
for col in categorical_cols:
  raw_df[col]=encoder.fit_transform(raw_df[col].values.reshape(1,-1)).toarray().reshape(-1)
train_inputs[categorical_cols]=raw_df[categorical_cols]
val_inputs[categorical_cols]=raw_df[categorical_cols]
test_inputs[categorical_cols]=raw_df[categorical_cols]
```

```python
train_inputs.info()
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=150)
model.fit(train_inputs,train_target)
```

```python
val_pred=model.predict(val_inputs)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(val_target,val_pred)
score
```

```python
test_pred=model.predict(test_inputs)
score=accuracy_score(test_target,test_pred)
score
```
