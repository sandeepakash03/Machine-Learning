```python
import opendatasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
```

```python
opendatasets.download('https://www.kaggle.com/datasets/jsphyg/weather-dataset-rattle-package')
```

```python
weather_df=pd.read_csv('weather-dataset-rattle-package/weatherAUS.csv')
weather_df.head()
```

```python
weather_df.describe()
```

```python
weather_df.info()
```

```python
#changing the date object into datetime datatype
weather_df.Date=pd.to_datetime(weather_df.Date)
```

```python
weather_df.info()
```

```python
numerical_cols=weather_df.select_dtypes(np.number).columns.tolist()
categorical_cols=weather_df.select_dtypes('object').columns.tolist()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='mean')
weather_df[numerical_cols]=imputer.fit_transform(weather_df[numerical_cols])
```

```python
weather_df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='most_frequent')
weather_df[categorical_cols]=imputer.fit_transform(weather_df[categorical_cols])
```

```python
weather_df.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
scaler=StandardScaler()
weather_df[numerical_cols]=scaler.fit_transform(weather_df[numerical_cols])
```

```python
encoder=OneHotEncoder(handle_unknown='ignore')
```

```python
for category in categorical_cols:
    weather_df[category]=encoder.fit_transform(weather_df[category].values.reshape(1,-1)).toarray().reshape(-1)
```

```python
weather_df
```

```python
X=weather_df.iloc[:,1:-1]
y=weather_df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_jobs=-1,random_state=42,n_estimators=500,max_features=7,max_depth=55)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.predict_proba(X_test)
```

```python
model.estimators_[0]
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(5,10))
plot_tree(model.estimators_[20],max_depth=2,feature_names=X_train.columns,filled=True,rounded=True)
```
