```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris_df=pd.read_csv('/kaggle/input/iris/Iris.csv')
iris_df.head()
```

```python
iris_df.describe()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(iris_df,x='PetalLengthCm',y='PetalWidthCm',hue='Species')
plt.title('Clustering would be look like')
plt.plot()
```

```python
sns.scatterplot(iris_df,x='PetalLengthCm',y='PetalWidthCm',)
plt.title('without clustering the data points would looks like')
plt.plot()
```

```python
X=iris_df.drop(['Id','Species'],axis=1)
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=3,random_state=42)
```

```python
model.fit(X)
```

```python
pred=model.predict(X)
```

```python
model.predict(X)
```

```python
model.inertia_
```

```python
model.cluster_centers_
```

```python
sns.scatterplot(iris_df,x='PetalLengthCm',y='PetalWidthCm',hue=pred,alpha=0.6,)
plt.plot(model.cluster_centers_[:,2],model.cluster_centers_[:,-1],'or')
```
